# Crew AI

- CrewAI Enterprise
    - A multi-agent platform for deploying, running and monitoring Agentic AI
- CrewAI UI Studio
    - A no-code / lo-code product for creating multi-agent solutions
- CrewAI open-source framework
    - Orchestrate high performing AI agents with case and scale


### CrewAI Crews
- Autonomous solutions with AI teams of Agents with different roles
- Choose crew when: You need autonomous problem-solving, creative collaboration, or exploratory tasks

### CrewAI Flows
- Structured automations by dividing complex tasks into precise workflows
- Choose Flows when: you require deterministic outcomes, auditability, or precise control over execution


## Core Concepts
- **Agent**: an autonomous unit, with an LLM, a role, a goal, a backstory, memory, tools
- **Task**: a specific assignment to be carried out, with description, expected output, agent
- **Crew**: a team of Agents and Tasks; either
    - *Sequential*: run tasks in order they are defined
    - *Hierarchical*: use a Manager LLM to assign

Crew is lightweight, but somewhat more opinionated than OpenAI Agents SDK - more terminology, marginally more prescriptive and with an ability to get much more prescriptive

* Agents and tasks can be created by code, setting the backstory, description, expected output, etc. Or you can define each in a YAML file that is provided when you create the code
* We need to write the code in a crew.py file

Create a new project with
     `crewai create crew my_crew` or `crewai create flow my_crew`

run it with `crewai run`

## Summarization
1. Create the project with: `crewai create crew my_project`
2. Fill in the config ymal files to define **Agents** and **Tasks**
3. Complete the crew.py module to create the **Agents**, **Tasks**, and **Crew**, referencing the config
4. Update *main.py* to set any inputs
5. Run with: `crewai run`

## Memory

1. *Short-Term Memory*: Temporarily stores recent interactions and outcomes using RAG, enabling agents to access relevant information during the current executions
2. *Long-Term Memory*: Preserves valuable insights and learnings, building knowledge over time
3. *Entity Memory*: Information about people, places, and concepts encountered during tasks, facilitating deeper understanding and relationship mapping. Uses RAG for storing entity information
4. *Contextual Memory*: Maintaining the context of interaction by combining all the above
5. *User Memory*: Stores user-specific information and preferences, enhancing personalization and user experience (this is up to us to manage and include in prompts)

* You can ask crew to write a code and run it in a docker container by adding `allow_code_execution=True` and `code_execution_mode='safe'`
